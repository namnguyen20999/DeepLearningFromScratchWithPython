```python
from IPython.display import HTML
from itertools import groupby

FOLDER_COLORS = {"core": "#6366f1", "chain_rule": "#d97706", "matrix_ops": "#16a34a"}

rows = []
for folder, group in groupby(FILES, key=lambda f: f["folder"]):
    fc = FOLDER_COLORS.get(folder, "#334155")
    rows.append(f'<tr><td colspan="5" style="background:{fc}22;color:{fc};'
                f'font-weight:bold;font-size:12px;padding:4px 10px">'
                f'📁 {folder}/</td></tr>')
    for f in group:
        fns  = "<br>".join(f"<code>{n}</code> — {d}" for n, d in f["functions"])
        imps = "<br>".join(f["imports"]) if f["imports"] else "—"
        rows.append(f"""
          <tr>
            <td style="white-space:nowrap;font-weight:bold;color:{f['color']};padding-left:20px">
              {f['file'].split('/')[1]}</td>
            <td style="white-space:nowrap">{f['kind']}</td>
            <td>{fns}</td>
            <td style="font-size:0.85em;color:#555">{imps}</td>
            <td style="font-size:0.85em;color:#555">{f['note']}</td>
          </tr>""")

nb_rows = "".join(
    f'<tr><td style="font-weight:bold;padding-left:20px">{nb.split("/")[1]}</td>'
    f'<td colspan="4">{desc}</td></tr>'
    for nb, desc in NOTEBOOKS
)

html = f"""
<style>
  table.ref {{ border-collapse:collapse; font-family:monospace; font-size:13px; width:100% }}
  table.ref th {{ background:#1e293b; color:#f8fafc; padding:6px 10px; text-align:left }}
  table.ref td {{ padding:5px 10px; border-bottom:1px solid #e2e8f0; vertical-align:top }}
  table.ref tr:hover td {{ background:#f8fafc }}
</style>
<table class="ref">
  <thead>
    <tr><th>File</th><th>Kind</th><th>Functions / Symbols</th>
        <th>Imports from</th><th>Notes / How to run</th></tr>
  </thead>
  <tbody>
    {''.join(rows)}
    <tr><td colspan="5" style="background:#33415522;color:#334155;font-weight:bold;
        font-size:12px;padding:4px 10px">📓 notebooks/</td></tr>
    {nb_rows}
  </tbody>
</table>
"""
HTML(html)
```
